```
# importing libraries
import pandas as pd

# Creating the Class 'State' to model each state in the problem
class State:

    # initializing the state class attributes
    def __init__(self, cannibalLeft, missionaryLeft, boat, \
                 cannibalRight, missionaryRight, action, h, g):
        self.cannibalLeft = cannibalLeft
        self.missionaryLeft = missionaryLeft
        self.boat = boat
        self.cannibalRight = cannibalRight
        self.missionaryRight = missionaryRight
        self.action = action
        self.h = h
        self.g = g
        self.f = h + g
        self.parent = None

    # method to check whether the state is the goal state or not
    def check_goal_state(self):
        if self.cannibalLeft == 0 and self.missionaryLeft == 0:
            return True
        else:
            return False

    # method to check whether the move is valid or not
    def check_valid_move(self):
        if self.missionaryLeft >= 0 and self.missionaryRight >= 0 \
                and self.cannibalLeft >= 0 and self.cannibalRight >= 0 \
                and (self.missionaryLeft == 0 or self.missionaryLeft >= \
                     self.cannibalLeft) and (self.missionaryRight == 0 or \
                                    self.missionaryRight >= self.cannibalRight):
            return True
        else:
            return False
```

```
# function to generate the leaves of the current node
# return the list of the current state's children obeying valid moves

def find_children(current_state):
    children = []

    # specifying what to do if the boat is on the left side
    # also checks the move validity
    if current_state.boat == 'left':

        # moving two Missionories to right side
        new_state = State(current_state.cannibalLeft, current_state.missionaryLeft - 2, 'right',
                          current_state.cannibalRight, current_state.missionaryRight + 2,
                          "M,M ----->",
                          current_state.cannibalLeft + current_state.missionaryLeft - 1, \
                          current_state.g + 1)
        if new_state.check_valid_move():
            new_state.parent = current_state
            children.append(new_state)

        # moving two Cannibals to right side
        new_state = State(current_state.cannibalLeft - 2, current_state.missionaryLeft, 'right',
                          current_state.cannibalRight + 2, current_state.missionaryRight,
                          "C,C ----->",
                          current_state.cannibalLeft + current_state.missionaryLeft - 1, \
                          current_state.g + 1)
        if new_state.check_valid_move():
            new_state.parent = current_state
            children.append(new_state)

        # moving one M and one C to right side
        new_state = State(current_state.cannibalLeft - 1, current_state.missionaryLeft - 1, 'right',
                          current_state.cannibalRight + 1, current_state.missionaryRight + 1,
                          "M,C ----->",
                          current_state.cannibalLeft + current_state.missionaryLeft - 1, \
                          current_state.g + 1)
        if new_state.check_valid_move():
            new_state.parent = current_state
            children.append(new_state)

        # moving one M to right side
        new_state = State(current_state.cannibalLeft, current_state.missionaryLeft - 1, 'right',
                          current_state.cannibalRight, current_state.missionaryRight + 1,
                          "M ----->",
                          current_state.cannibalLeft + current_state.missionaryLeft - 1, \
                          current_state.g + 1)
        if new_state.check_valid_move():
            new_state.parent = current_state
            children.append(new_state)

        # moving one C to right side
        new_state = State(current_state.cannibalLeft - 1, current_state.missionaryLeft, 'right',
                          current_state.cannibalRight + 1, current_state.missionaryRight,
                          "C ----->",
                          current_state.cannibalLeft + current_state.missionaryLeft - 1, \
                          current_state.g + 1)
        if new_state.check_valid_move():
            new_state.parent = current_state
            children.append(new_state)

    # specifying what to do if the boat is on the right side
    else:

        # moving two missionaries to left side
        new_state = State(current_state.cannibalLeft, current_state.missionaryLeft + 2, 'left',
                          current_state.cannibalRight, current_state.missionaryRight - 2,
                          "<----- M,M",
                          current_state.cannibalLeft + current_state.missionaryLeft - 1, \
                          current_state.g + 1)
        if new_state.check_valid_move():
            new_state.parent = current_state
            children.append(new_state)

        # moving two cannibals to left side
        new_state = State(current_state.cannibalLeft + 2, current_state.missionaryLeft, 'left',
                          current_state.cannibalRight - 2, current_state.missionaryRight,
                          "<----- C,C",
                          current_state.cannibalLeft + current_state.missionaryLeft - 1, \
                          current_state.g + 1)
        if new_state.check_valid_move():
            new_state.parent = current_state
            children.append(new_state)

        # moving one M and one C to left side
        new_state = State(current_state.cannibalLeft + 1, current_state.missionaryLeft + 1, 'left',
                          current_state.cannibalRight - 1, current_state.missionaryRight - 1,
                          "<----- M,C",
                          current_state.cannibalLeft + current_state.missionaryLeft - 1, \
                          current_state.g + 1)
        if new_state.check_valid_move():
            new_state.parent = current_state
            children.append(new_state)

        # moving one M to left side
        new_state = State(current_state.cannibalLeft, current_state.missionaryLeft + 1, 'left',
                          current_state.cannibalRight, current_state.missionaryRight - 1,
                          "<----- M",
                          current_state.cannibalLeft + current_state.missionaryLeft - 1, \
                          current_state.g + 1)
        if new_state.check_valid_move():
            new_state.parent = current_state
            children.append(new_state)

        # moving one C to left side
        new_state = State(current_state.cannibalLeft + 1, current_state.missionaryLeft, 'left',
                          current_state.cannibalRight - 1, current_state.missionaryRight,
                          "<----- C",
                          current_state.cannibalLeft + current_state.missionaryLeft - 1, \
                          current_state.g + 1)
        if new_state.check_valid_move():
            new_state.parent = current_state
            children.append(new_state)

    return children
```

```
# function to find out the solution using a* algorithm

def find_solution():

    # initial state as an instance of the class State
    # 3M, 3C at the left side and the boat is also on the left side
    # g = 0, h = 3 + 3 - 1 = 5
    # heuristic: number of people on initial side - 1
    initial_state = State(3, 3, 'left', 0, 0, "", 5, 0)

    # checking if the initial state itselft is the goal state
    if initial_state.check_goal_state():
        return initial_state

    # initializing open list and closed (visited) list
    open_list = list()
    visited = set()

    # adding the initial state into the open list
    open_list.append(initial_state)

    while open_list:

        # initializing the cost list for f
        # f = g + h
        costs = []

        # finding the cost of each node in the open list
        for node in open_list:
            costs.append(node.h + node.g)

        # finidng the state with minimum cost
        index = costs.index(min(costs))
        state = open_list.pop(index)

        # checking whether the minimum cost state is the goal state or not
        if state.check_goal_state():
            return state

        # adding the state to the closed (visited) list
        visited.add(state)

        # finding the leaves of the node (children)
        children = find_children(state)

        # setting the parameters g and parent for the children
        # also checks whether it was already visited or not

        for child in children:
          if (child in visited) and (state.f < child.f):
              child.g = state.g + 1
              child.h = child.missionaryLeft + child.cannibalLeft - 1
              child.parent = state
          elif (child in open_list) and (state.f < child.f):
              child.g = state.g + 1
              child.h = child.missionaryLeft + child.cannibalLeft - 1
              child.parent = state
          else:
              open_list.append(child)
              child.g = state.g + 1
              child.h = child.missionaryLeft + child.cannibalLeft - 1
    return None
```

```
# function to find the path of the solution
# backtrack from the goal state to the initial state

def find_path(solution):

    # initializing the path list
    path = []
    path.append(solution)
    parent = solution.parent

    # backtrack
    # trace the parent
    while parent:
        path.append(parent)
        parent = parent.parent

    # setting the needed parameters to display the solution and path
    for i in range(len(path)):
        state = path[len(path) - i - 1]
        step_c.append(i)
        step.append(state.action)
        l_c.append(state.cannibalLeft)
        l_m.append(state.missionaryLeft)
        boat.append(state.boat)
        r_c.append(state.cannibalRight)
        r_m.append(state.missionaryRight)
        g.append(state.g)
        h.append(state.h)
        f.append(state.f)
```

```
# initializing the needed parameters as empty lists
step_c, step, boat, l_c, l_m, r_c, r_m, g, h, f = ([] for i in range(10))

# function call to find the solution and its path
find_path(find_solution())

# display the solution and path as Pandas Dataframe
data = list(zip(step_c, step, boat, l_c, l_m, r_c, r_m, g, h, f))
df = pd.DataFrame(data,
                  columns=['step', 'movement', 'boat pos', 'left C', 'left M', \
                           'right C', 'right M', 'g', 'h', 'f'])
print(df.to_string(index=False))
```
